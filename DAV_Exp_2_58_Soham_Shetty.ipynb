```r
import pandas as pd

df = pd.read_csv('/datalab/placement.csv')

print(df.info())
print(df.describe())
```

```r
import seaborn as sns
sns.histplot(df['cgpa'], kde=True)
```

```r
import matplotlib.pyplot as plt
from scipy import stats

stats.probplot(df['cgpa'], plot=plt)
plt.title('Q-Q Plot of Dependent Variable')
plt.show()
```

```r
sns.regplot(x='placement_exam_marks', y='cgpa', data=df, lowess=True, line_kws={'color': 'red'})
plt.title('Scatterplot with Lowess Fit Line')
```

```r
from sklearn.linear_model import LinearRegression

X = df[['placement_exam_marks']]
y = df['cgpa']

model = LinearRegression()
model.fit(X, y)

# Display coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```

```r
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='green', linewidth=2)
plt.xlabel('placement_exam_marks')
plt.ylabel('cgpa')
plt.show()
```

# R Simple Linear Regression


```r
library(tidyverse)

df <- read.csv('/datalab/placement.csv')

str(df)
summary(df)
```

```r
hist(df$cgpa, main="Histogram", xlab="cgpa")
```

```r
plot(df$placement_exam_marks, df$cgpa, main="Scatterplot", xlab="placement_exam_marks", ylab="cgpa")
```

```r
model <- lm(cgpa ~ placement_exam_marks, data=df)


summary(model)


plot(df$placement_exam_marks, df$cgpa, main="Linear Regression", xlab="placement_exam_marks", ylab="cgpa")
abline(model, col='red', lwd=2)
```
